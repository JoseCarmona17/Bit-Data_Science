```python
import pandas as pd

df = pd.read_csv('youtube-top-100-songs-2025.csv')
df.head()
```

```python
# info
df.info()
```

```python
# Describe
df.describe()
```

```python
#Genere un nuevo dataset aplicando un filtro sobre los datos originales

# Filtrar canciones con más de 100 millones de vistas
df_filtrado = df[df['view_count'] > 100_000_000]

# Ver las primeras filas del nuevo dataset
df_filtrado.head()
```

```python
# Genere un nuevo dataset aplicando un ordenamiento sobre los datos originales

# Ordenar por vistas de mayor a menor
df_ordenado = df.sort_values(by='view_count', ascending=False)

# Ver las primeras filas del nuevo dataset ordenado
df_ordenado.head()
```
